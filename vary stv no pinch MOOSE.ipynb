```python
################################################################################################ At this point, the physical problem has been solved and all data has been mined. We can now # # move the files from projects/ferret/ to the /calculations/ folder                           ################################################################################################import numpy as npimport matplotlib.pyplot as pltfor k in range(0,121):        data_SigmaVolume = np.genfromtxt('/home/john/projects/calculations/core_shell/vary_stv_120_bg_fixedmesh/MeshData-'+str(k)+'-0.csv',delimiter=',', dtype = float)    # stresses    sxx = [row[0] for row in data_SigmaVolume] #generate vectors of data  can forloop this for all columns of data    syy = [row[1] for row in data_SigmaVolume]    szz = [row[2] for row in data_SigmaVolume]    # strains    exx = [row[3] for row in data_SigmaVolume]    eyy = [row[4] for row in data_SigmaVolume]    ezz = [row[5] for row in data_SigmaVolume]    # volume    vol = [row[12] for row in data_SigmaVolume]        nbins = 100    Hxx, xedgesxx, yedgesxx = np.histogram2d(sxx[1:len(sxx)],vol[1:len(vol)],bins=nbins)    Hyy, xedgesyy, yedgesyy = np.histogram2d(syy[1:len(szz)],vol[1:len(vol)],bins=nbins)    Hzz, xedgeszz, yedgeszz = np.histogram2d(szz[1:len(szz)],vol[1:len(vol)],bins=nbins)    Hxx = np.rot90(Hxx)    Hxx = np.flipud(Hxx)    Hyy = np.rot90(Hyy)    Hyy = np.flipud(Hyy)    Hzz = np.rot90(Hzz)    Hzz = np.flipud(Hzz)#here we are summing V[j]=Table[Hxx[k,j]v[k],{k,all}]    vxxj = np.einsum('ij,i',Hxx,(yedgesxx[1:nbins+1]+yedgesxx[0:nbins])/2)    vyyj = np.einsum('ij,i',Hyy,(yedgesyy[1:nbins+1]+yedgesyy[0:nbins])/2)    vzzj = np.einsum('ij,i',Hzz,(yedgeszz[1:nbins+1]+yedgeszz[0:nbins])/2)#note we have to assume the value is in between the bin edges    Vxxj = vxxj/sum(vxxj)*100    Vyyj = vyyj/sum(vyyj)*100    Vzzj = vzzj/sum(vzzj)*100    x_xx = (xedgesxx[1:nbins+1]+xedgesxx[0:nbins])/2    x_yy = (xedgesyy[1:nbins+1]+xedgesyy[0:nbins])/2    x_zz = (xedgeszz[1:nbins+1]+xedgeszz[0:nbins])/2    fig = plt.figure()    ax = fig.add_subplot(111, axisbg = 'w')    ax.plot(x_xx,Vxxj,'r',linewidth=1.8, label = '$\sigma_{xx}$')    ax.plot(x_yy,Vyyj,'g',linewidth=1.8, label = '$\sigma_{yy}$')    ax.plot(x_zz,Vzzj,'b',linewidth=1.8, label = '$\sigma_{zz}$')    plt.rc('text', usetex=True)    plt.rc('font', family='serif')        plt.grid(True,color = 'k')    #plt.title('$%$ Shell volume fraction of $sigma_{jj}$ at $p$ =' + str(pressure_vec[k]) +' N/nm^2')    plt.ylabel('Shell volume fraction [$\%$] ')    plt.xlabel('$\sigma_{jj}$')    plt.legend(labelspacing = 0.6,loc=4)    plt.savefig("/home/john/projects/calculations/core_shell/vary_stv_120_bg_fixedmesh/histograms_stv_test-"+str(k)+".pdf")    #plt.show()
```
