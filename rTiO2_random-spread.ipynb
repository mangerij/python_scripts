```python
# numpy genfromtxt crashes when loading large files, have to use an iterative loading scheme# Found on stackexchange with some interesting plots on memory. We are essentially using this sceme above with the pvpython scriptsimport numpy as npdef iter_loadtxt(filename, delimiter=',', skiprows=1, dtype=float):    def iter_func():        with open(filename, 'r') as infile:            for _ in range(skiprows):                next(infile)            for line in infile:                line = line.rstrip().split(delimiter)                for item in line:                    yield dtype(item)        iter_loadtxt.rowlength = len(line)    data = np.fromiter(iter_func(), dtype=dtype)    data = data.reshape((-1, iter_loadtxt.rowlength))    return data
```
